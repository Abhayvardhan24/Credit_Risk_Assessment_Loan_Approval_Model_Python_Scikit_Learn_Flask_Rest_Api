```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Loan_approval_data_2025.csv')
```

```python
df['occupation_status'].unique()
```

```python
def occu(x):
    if x == 'Employed':
        return 0
    if x == 'Student' :
        return 1
    if x == 'Self-Employed':
        return 2
    else:
        return x
    
```

```python
df['occupation_status'] = df['occupation_status'].apply(occu)
```

```python
df['product_type'].unique()
```

```python
def pro(x):
    if x == 'Credit Card':
        return 1
    if x == 'Personal Loan':
        return 2
    if x == 'Line of Credit':
        return 3
    else:
        return x
```

```python
df['product_type'] = df['product_type'].apply(pro)
```

```python
df['loan_intent'].unique()
```

```python
def lon_intent(x):
    if x == 'Business':
        return 1
    if x == 'Home Improvement':
        return 2
    if x == 'Debt Consolidation':
        return 3
    if x == 'Education':
        return 4
    if x == 'Personal':
        return 5
    if x == 'Medical':
        return 6
    else:
        return x
```

```python
df['loan_intent'] = df['loan_intent'].apply(lon_intent)
```

```python
df
```

```python
df.drop(columns=['customer_id'], inplace=True)
```

```python
def con(x):
    if x == True:
        return 1
    if x == False:
        return 0
    else:
        return x
```

```python
df['loan_approved'] = df['loan_approved'].apply(con)
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.info()
```

```python
q1 = df.quantile(0.25)
q3 = df.quantile(0.75)
iqr = q3 - q1
upper = q3+1.5 *(iqr)
lower = q1 - 1.5 * (iqr)
```

```python
data = df[(df <= upper) & (df >= lower)]
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.isnull().sum()
```

```python
data.info()
```

```python
data.to_csv('Model_data_xyz.xlsx', index = False)
```
